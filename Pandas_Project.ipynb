```python

import pandas as pd
df=pd.read_csv("/content/uncleaned_airline_dataset_20_columns (1) (1).csv")
df
```

```python
#Remove rows where Customer_ID is missing.
df = df.dropna(subset=['Customer_ID'])
df
```

```python
#Verify that missing values are handled.
df.isnull().sum()
```

```python
#Remove leading and trailing spaces from all string columns.
df = df.apply(lambda col: col.str.strip() if col.dtype == 'object' else col)
df
```

```python
#Convert Meal_Option values to lowercase.
df['Meal_Option'] = df['Meal_Option'].str.lower()
df
```

```python
#Convert Ticket_Type values to proper format (Refundable / Non-Refundable).
df['Ticket_Type'] = df['Ticket_Type'].str.strip().str.lower()

df['Ticket_Type'] = df['Ticket_Type'].replace({
    'refundable': 'Refundable',
    'non refundable': 'Non-Refundable',
    'non-refundable': 'Non-Refundable'
})
df
```

```python
#Standardize Checkin_Status values.
df['Checkin_Status'] = df['Checkin_Status'].str.strip().str.lower()

df['Checkin_Status'] = df['Checkin_Status'].replace({
    'checked in': 'Checked-In',
    'check-in': 'Checked-In',
    'yes': 'Checked-In',
    'no': 'Not Checked-In',
    'not checked in': 'Not Checked-In'
})

df
```

```python
#Standardize Airport_Terminal values to T1, T2, T3 format.
df['Checkin_Status'] = df['Checkin_Status'].str.strip().str.lower()

df['Checkin_Status'] = df['Checkin_Status'].replace({
    'checked in': 'Checked-In',
    'check-in': 'Checked-In',
    'yes': 'Checked-In',
    'no': 'Not Checked-In',
    'not checked in': 'Not Checked-In'
})

df
```

```python
#Convert Passenger_Rating column to numeric datatype.
df['Passenger_Rating'] = pd.to_numeric(df['Passenger_Rating'])
df
```

```python
#Convert Baggage_kg values by removing "kg" text.
df['Baggage_kg'] = df['Baggage_kg'].str.replace('kg', '', regex=False)
df
```

```python
#Convert Delay_Minutes column into integer datatype.
df['Delay_Minutes'] = pd.to_numeric(df['Delay_Minutes'], errors='coerce').astype('Int64')
df
```

```python
# Replace "unknown" values with NaN.
import numpy as np
df.replace('unknown', np.nan, inplace=True)
df
```
